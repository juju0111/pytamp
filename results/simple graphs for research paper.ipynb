```python
from matplotlib import pyplot as plt
import numpy as np

%matplotlib inline
```

### Simple Plot

```python
x_list = np.linspace(0.,4.,10)
y_list = (x_list-1.)**2 + 2.

plt.plot(x_list, y_list,'-o', lw=3, ms=15,label="Plot Name") # lw 선 굵기 조절 (논문에 들어갔을 때 보일 정도로 조절), ms 마커사이즈 조절 (논문에 들어갔을 때 보일 정도로 조절)

fontname = 'sans-serif' # font
plt.xlim([0., 4.])
plt.ylim([1., 12.])
plt.xticks(fontsize=18, fontname=fontname) # x축 눈금 font, fontsize 조절
plt.yticks(fontsize=18, fontname=fontname) # y축 눈금 font, fontsize 조절
plt.xlabel('X-Axis',fontsize=28, fontname=fontname) # x축 이름 조절
plt.ylabel('Y-Axis',fontsize=28, fontname=fontname) # y축 이름 조절
plt.legend(loc="upper left",prop={'size':22,'family':fontname}) # legend 위치, font, fontsize 조절
plt.show()
```

### Mean Variance Plot

```python
x_list = np.linspace(0.,4.,10)
mean_list = (x_list-1.)**2 + 2.
std_list = 2.*np.ones_like(x_list)

plt.fill_between(x_list, mean_list-std_list,mean_list+std_list,alpha=0.1,)#label="STD Plot") # 보통 std는 label 달지 않음.
plt.plot(x_list, mean_list,'-o', lw=3, ms=15,label="Mean Plot")

fontname = 'sans-serif'
plt.xlim([0., 4.])
plt.ylim([1., 12.])
plt.xticks(fontsize=18, fontname=fontname)
plt.yticks(fontsize=18, fontname=fontname)
plt.xlabel('X-Axis',fontsize=28, fontname=fontname)
plt.ylabel('Y-Axis',fontsize=28, fontname=fontname)
plt.legend(loc="upper left",prop={'size':22,'family':fontname})
plt.show()
```
